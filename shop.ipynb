```python
import pandas as pd
import matplotlib.pyplot as plt

# Baca data dari file CSV
df1 = pd.read_csv('shopping_behavior_updated.csv')
df2 = pd.read_csv('shopping_trends_updated.csv')
df3 = pd.read_csv('shopping_trends.csv')

# Gabungkan data dari ketiga file
combined_data = pd.concat([df1, df2, df3])

# Menghitung total pembelian dari masing-masing file
total_purchase_file1 = df1['Purchase Amount (USD)'].sum()
total_purchase_file2 = df2['Purchase Amount (USD)'].sum()
total_purchase_file3 = df3['Purchase Amount (USD)'].sum()

# Data untuk grafik lingkaran
labels = ['File 1', 'File 2', 'File 3']
sizes = [total_purchase_file1, total_purchase_file2, total_purchase_file3]
explode = (0.1, 0, 0)  # Jika Anda ingin "memecah" sektor pertama

# Buat grafik lingkaran
plt.figure(figsize=(8, 8))
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=140, explode=explode)
plt.axis('equal')  # Pastikan lingkaran terlihat sebagai lingkaran

plt.title('Perbandingan Total Pembelian dari Tiga File CSV')
plt.show()
```
